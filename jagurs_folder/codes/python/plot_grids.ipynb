```python
import pygmt

# Load the grid data
# Create a figure for each grid
k=0
r=[r1,r2,r3,r4,r5]
for i in grids:
    k=k+1
    # Read the grid data
    fig = pygmt.Figure()
    fig.basemap(region=r[k-1], projection="M6i",frame="af")
    if k==1:
        cmap=pygmt.makecpt(cmap='geo',series=[-6000,6000,100],continuous=True,reverse=True)
    elif k==2:
        cmap=pygmt.makecpt(cmap='geo',series=[-6000,6000,100],continuous=True,reverse=True)
    elif k==3:
        cmap=pygmt.makecpt(cmap='geo',series=[-2000,2000,100],continuous=True,reverse=True)
    elif k==4:
        cmap=pygmt.makecpt(cmap='geo',series=[-300,300,25],continuous=True,reverse=True)
    elif k==5:
        cmap=pygmt.makecpt(cmap='geo',series=[-300,300,25],continuous=True,reverse=True)

    fig.grdimage(grid=i, cmap=cmap)
    fig.colorbar(frame=["x+lProfundidad", "y+lm"])
    # Set the region and projection
    if k==1:
        fig.plot(data=rectangle, style="r+s", pen="2p,red",projection='M6i')
        fig.plot(data=rectangle2, style="r+s", pen="2p,red",projection='M6i')
        fig.plot(data=rectangle3, style="r+s", pen="2p,red",projection='M6i')
        fig.plot(data=rectangle4, style="r+s", pen="2p,red",projection='M6i')
    if k==2:
        fig.plot(data=rectangle2, style="r+s", pen="2p,red",projection='M6i')
        fig.plot(data=rectangle3, style="r+s", pen="2p,red",projection='M6i')
        fig.plot(data=rectangle4, style="r+s", pen="2p,red",projection='M6i')
    if k==3:
        fig.plot(data=rectangle3, style="r+s", pen="2p,red",projection='M6i')
        fig.plot(data=rectangle4, style="r+s", pen="2p,red",projection='M6i')
    if k==4:
        fig.plot(data=rectangle4, style="r+s", pen="2p,red",projection='M6i')
    # Plot the grid

    # Add coastlines
    # Save the figure
    fig.savefig(f"figure{k+1}.png")

    # Show the figure
    fig.show()
    
```
